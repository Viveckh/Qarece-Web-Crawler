```python
import pandas as pd
```

```python
products = pd.read_csv('Product copy.csv')
```

```python
products.head()
```

```python
products = products.sort_values(by='wp_product_type')
```

```python
total = len(products)

batch = 25000
```

```python
all_dfs = []
start = 0
end = batch
sum = 0
while start < total:
    df = pd.DataFrame(products[start:end])
    all_dfs.append(df)
    sum = sum + len(df)
    start = end
    end = start + batch
    
    if end > total:
        end = total
```

```python
for index, d in enumerate(all_dfs):
    d.to_csv('product_'+str(index+1))
```
