```python
# Imports
from bs4 import BeautifulSoup
import requests
```

```python
# Initial Variables, set pageSize to 10 always
base_url = "https://www.sephora.com/shop/value-sets?sortBy=PRICE_HIGH_TO_LOW&pageSize=10"
current_page = 1
product_url_list = []
```

```python
# Function that accepts the search page url to parse along with current page
# The paging was necessary since search pages otherwise have a lazy loading in place which makes parsing tricky
def get_product_urls(base_url, current_page):
    response = requests.get(base_url + "&currentPage=" + str(current_page))
    soup = BeautifulSoup(response.content)

    product_urls = []
    for link in soup.select('a[data-comp~="ProductItem"]'):
        product_urls.append("https://www.sephora.com" + link.get('href'))
    return product_urls
```

```python
# Begin parsing and keep proceeding to next page until you can no longer retrieve products
while True:
    new_product_urls = get_product_urls(base_url, current_page)
    product_url_list = product_url_list + new_product_urls
    if len(new_product_urls) > 0:
        current_page = current_page + 1
    else:
        break

print(len(product_url_list))
product_url_list
```

```python
### Writing to file
with open('urls.txt', 'w') as f:
    for item in product_url_list:
        f.write("%s\n" % item)
```
